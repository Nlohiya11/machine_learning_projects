```python
def range(x):
    temp = x.split('-')
    temp1 = x.split('*')
    if len(temp)== 2:
        return (float(temp[0])+float(temp[1]))/2
    if len(temp1)== 2:
            return (float(temp1[0])*float(temp1[1]))
    try:
        return float(x)
    except:
        return None
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state = 2)
```

```python
from sklearn.linear_model import LinearRegression , Lasso, Ridge
x_train,y_train = df.drop(columns=['price'],axis =1),df['price']
reg = LinearRegression()
```

```python
columns_trans = make_column_transformer((OneHotEncoder(sparse_output = False),['location']),remainder='passthrough')
```

```python
scaler = StandardScaler()
```

```python
pipe = make_pipeline(columns_trans,scaler,reg)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_lr = pipe.predict(x_test)
```

```python
r2_score(y_test, y_pred_lr)
```

```python
las = Lasso()
```

```python
pipe_las = make_pipeline(columns_trans,scaler,las)
```

```python
pipe_las.fit(x_train,y_train)
```

```python
y_pred_las = pipe_las.predict(x_test)
```

```python
r2_score(y_test, y_pred_las)
```

```python
rid = Ridge()
```

```python
pipe_rid = make_pipeline(columns_trans,scaler,rid)
pipe_rid.fit(x_train,y_train)
y_pred_rid = pipe_rid.predict(x_test)
r2_score(y_test, y_pred_rid)
```
