```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('MagicBricks.csv',encoding="ISO-8859-1")
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df = df.drop(columns=["Furnishing","Status","Transaction","Type","Per_Sqft"])
df.sample(5)
```

```python
df.describe()
```

```python
df.info()
```

```python
df[df['Bathroom'].isnull()]
```

```python
df['Bathroom']=df['Bathroom'].fillna(0)
```

```python
df['Parking']=df['Parking'].fillna(0)
```

```python
df['Per_sqrt'] = (np.round(df['Price']/df['Area'],2))
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df['Price'] = df['Price']/100000
```

```python
df.describe()
```

```python
df = df.drop(columns=["Per_sqrt","Parking"])
df.sample(5)
```

```python
y = df['Price']
x=df.drop(columns=['Price'],axis =1)
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state = 2)
```

```python
from sklearn.linear_model import LinearRegression , Lasso, Ridge
x_train,y_train = df.drop(columns=['Price'],axis =1),df['Price']
reg = LinearRegression()
```

```python
columns_trans = make_column_transformer((OneHotEncoder(sparse_output = False),['Locality']),remainder='passthrough')
scaler = StandardScaler()
pipe = make_pipeline(columns_trans,scaler,reg)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_lr = pipe.predict(x_test)
```

```python
r2_score(y_test, y_pred_lr)
```

```python
las = Lasso()
pipe_las = make_pipeline(columns_trans,scaler,las)
```

```python
pipe_las.fit(x_train,y_train)
```

```python
y_pred_las = pipe_las.predict(x_test)
r2_score(y_test, y_pred_las)
```

```python
rid = Ridge()
pipe_rid = make_pipeline(columns_trans,scaler,rid)
pipe_rid.fit(x_train,y_train)
```

```python
y_pred_rid = pipe_rid.predict(x_test)
r2_score(y_test, y_pred_rid)
```
