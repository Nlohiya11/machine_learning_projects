```python
from sklearn.model_selection import train_test_split
```

```python
airline_categoricalCol = ['airline','source_city','departure_time','destination_city','class']
dum_df = pd.get_dummies(df,columns = airline_categoricalCol,dtype=int)
dum_df.head(20)
```

```python
x = dum_df.drop(columns=['price'],axis=1)         
y = dum_df['price']  
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.25,random_state = 23,shuffle = True)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.metrics import r2_score 
import sklearn.metrics
```

```python
dtr = DecisionTreeRegressor(random_state=23,max_depth=None)
reg = LinearRegression()
les = Lasso()
rid = Ridge()
rfr = RandomForestRegressor()
```

```python
dtr.fit(x_train,y_train)
y_pred = dtr.predict(x_test)
r2_score(y_test,y_pred)
```

```python
reg.fit(x_train,y_train)
y_pred = reg.predict(x_test)
r2_score(y_test,y_pred)
```

```python
les.fit(x_train,y_train)
y_pred = les.predict(x_test)
r2_score(y_test,y_pred)
```

```python
rid.fit(x_train,y_train)
y_pred = rid.predict(x_test)
r2_score(y_test,y_pred)
```

```python
rfr.fit(x_train,y_train)
y_pred = rfr.predict(x_test)
r2_score(y_test,y_pred)
```
